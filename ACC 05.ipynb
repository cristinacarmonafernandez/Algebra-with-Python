### Pertenencia al radical (Proposici칩n 5.9)

```python
R.<x1,x2,x3> = PolynomialRing(GF(11),order='degrevlex'); R
```

```python
I = R.ideal([x1^11*x2^11, 
             x3^11, 
             x1^2 + x2*x3^2 + 7*x1*x2*x3, 
             4*x1*x2^2*x3 + 6*x2*x3^3 + 7*x1^2*x3]); I
```

```python
f = R(x1^2*x2^2 - 2*x1*x2*x3 + x3^2)
#f = R(x1^2*x2^2 + 2*x1*x2*x3 + x3^2)
```

```python
S.<x1,x2,x3,y> = PolynomialRing(GF(11),order='degrevlex'); S
```

```python
J = S.ideal([1 - S(f)*y] + [S(ele) for ele in I.gens()]); J
```

```python
J.groebner_basis()
```

```python
K = I.radical()
K
```

```python
G = I.groebner_basis();G
```

```python
f.reduce(G)
```

```python
H = K.groebner_basis()
H
```

```python
f.reduce(H)
```
